```python
# 3 Counting, the Dojo Way - Print integers from 1 to 100. If divisible by 5, print "Coding" instead. If divisible by 10, print "Coding Dojo."

for j in range(1,100,1):
  if j % 5 == 0 and j % 10 == 0:
    print("Coding Dojo")
  elif j % 5 == 0:
    print("Coding")
  else:
    print(j)
```
