# Exploring NumPy Arrays: Distance and Time

1) Import the numpy library

```python
import numpy as np
```

2) Create a numpy array called "distance" that includes the following distances (in meters): 100, 200, 400, 800, 1500, 3000, 5000, 10000

```python
distance = [100, 200, 400, 800, 1500, 3000, 5000, 10000] # in meters
distance_array = np.array(distance)
type(distance_array)
```

3) Use the np.min() and np.max() functions to find the minimum and maximum distances

```python
minimum_distance = distance_array.min()
maximum_distance = distance_array.max()

print(f"The minimum distance is {minimum_distance} meters, and the maximum distance is {maximum_distance} meters.")
```

4) To find the average distance, use the np.mean() function

```python
average_distance = distance_array.mean()
print("The mean distnace is: ", average_distance,"meters.")
```

5) use the np.sum() function to find the total distance

```python
total_distance = distance_array.sum()

print("The total distance is {} meters.".format(total_distance))
```

6) Another valuable function is the np.cumsum() function which outputs the cumulative sum...notice the output will be 100, 300, 700 as it adds each new distance

```python
cumulative_sum = distance_array.cumsum()

print("The cummulative sum of distances equals {} meters.".format(cumulative_sum))
```

7) To round all of the values in an array, use np.round()

```python
rounded_distance_array = distance_array.round()

print(rounded_distance_array)
```
